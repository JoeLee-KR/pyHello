### ch04. Collection - Dic, Set
### Dic { } : many used well case
- key-value, key base hashed store, no linear index, key is index of dictionary
- new key add, exist key update value

### Set { } : not used well, set is set in math, pure set
- no ref-indexing, no dup, pure set

```python
a = [0,1,2,3,4,5]
print('list a : ', a)
print(a[0],a[2])

b = {3:111, 2:11, 1:1, 0:0, 44:1111}
print('dictionary b :',b)
print(b[3])

a = {'korea':'seoul', 
     'canada':'ottawa',
     'usa':'wdc'}
print(a)
print(a['korea'])

#
a['japan'] = 'tokyo'
print(a)

a['korea'] = 'busan'
print(a)
```

### update()
- merge two dictionary, same key overwirte with added key-value
- pop method, del keyword

### access dictionary
- keys() 키만 반환
- values() 값만 반환
- items() kv의 tuple을 반환

```python
a = {'a':1, 'b':2, 'c':3}
b = {'a':33, 'd':44, 'e':55}
print(a,b)
a.update(b)
print(a,b)

# pop method
a.pop('b')
print(a)
# del keyword
del a['d']
print(a)
```

```python
print(a)
a.clear()
print(a)
```

```python
a = {'a':1, 'b':2, 'c':3}
c = [1,2,3,4,5,6,100]
print('b in a :', 'b' in a)
print('d in a :', 'd' in a)
print('2 in a :', 2 in a)
# list search is serial
# dictionary key search is hash index
print('100 in c :', 100 in c, '> list get is serial linear search')

# get(), nonstop pgm
print(a.get('d'), '> get method nonstop pgm')
# indexing, stop pgm
#print(a['d'])
if 'c' in a :
    print('if in check \'c\': ', a['c'])

# 'd' in a : is Crash error
```

```python
print(a)
print(a.keys())
print(a.values())
print(list(a.keys()))
print(list(a.values()))
print(list(a.items()))
```

```python
a = {1,2,3}
b = {2,3,4}
print(a.union(b))
print(b.union(a)) # 합집합
print(a.intersection(b)) # 교집합
print(a.difference(b)) # 차집합
print(a.issubset(b)) # 부분집합
```
