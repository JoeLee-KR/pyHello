### ch03. Collection - List [ ], Tuple ( )
- list: mutable / 생성 후 변경 가능 / [ ] 사용
- tuple: immutable / 생성 후 변경 불가 / contant성격이 강함

```python
a = [1,2,3,4,5]
print(a)
a = ['korea',1,2,3,4,5, [33,44]]
print(a)
```

### string split function

```python
a = 'hello world'
b = list(a)
print(b)

c = 'hello korea'
d = list(c)
print(d)

a = 'hello world nice weather'
b = str.split(a)
print(b)
```

### list indexing, slicing

```python
a = [1,2,3,4,5,6]
print(a[0], a[-1])

# a: list, string > immutable
b = 'hello world'
print(b[0])
#b[0] = 'j'
b = 'j' + b[1:]
print(b)

a[0] = 99
print(a)

# start;end:interval
print(a[2:3])
print(a[2:4])
print(a[:])
print(a[1:6:2])
```

### list member functions

```python
# append()
a = [1,2,3,4,5,6]
a.append(77)
print(a)

# extend
b = [7,8,9]
a.append(b)
print(a)
a.extend(b)
print(a)
a += [88,99]
print(a)

# insert
a.insert(1,444)

# remove with value
a.remove(77)
print(a)
a.remove([7,8,9])
print(a)

# pop with index
a.pop()
print(a)
d = a.pop(1)
print(a,d)

# index # with value
print(a.index(88))

# in keyword
print(88 in a)
print(0 in a)

# sort: itself list
# sorted: sort된 결과의 복사본 return
a.sort(reverse=True)
print(a)

d = sorted(a)
print(a,d)
```
